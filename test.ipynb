```python
# by뤼튼
from transformers import BertModel, BertTokenizer

# BERT 모델과 토크나이저 불러오기
model_name = 'bert-base-uncased'  # 예시로 'bert-base-uncased' 모델 사용
model = BertModel.from_pretrained(model_name)
tokenizer = BertTokenizer.from_pretrained(model_name)
```

```python
# 전처리된 텍스트를 BERT 입력 형식으로 변환하는 함수 정의
def preprocess_text(text):
    encoded_inputs = tokenizer.encode_plus(
        text,
        add_special_tokens=True,  # [CLS]와 [SEP] 토큰 추가
        max_length=512,  # 입력 최대 길이
        padding='max_length',  # 입력 길이를 맞추기 위한 패딩
        truncation=True,  # 입력 길이가 초과하는 경우 자르기
        return_attention_mask=True,  # 어텐션 마스크 생성
        return_tensors='pt'  # PyTorch 텐서로 반환
    )
    return encoded_inputs

# 예시 데이터 입력
text = "예시 문장입니다."

# 데이터 전처리 및 BERT 입력 형식으로 변환
inputs = preprocess_text(text)

# BERT 입력 형식으로 변환된 데이터 출력
print(inputs)
```
